```python
import xarray as xr
import datetime
```

```python
path = '/Users/jonniebarnsley/data/Observations/HadCRUT5/t2m/HadCRUT.5.0.1.0.median.nc'
ds = xr.open_dataset(path)
ds
```

```python
start = datetime.datetime(1850, 1, 16)
end = datetime.datetime(2022, 9, 16)
delta = datetime.timedelta(days=32)

current = start
dates=[current]
while current < end:
    current += delta
    current = datetime.datetime(current.year, current.month, 16) # adjust to 16th of the month
    dates.append(current)

ds = ds.assign_coords(time=dates)
ds
```
